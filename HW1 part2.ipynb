計算 Kobe「上半場得分多於下半場」且命中率最低的前 3 名場次，列出日期、對手、上
下半場得分差、該場得分，按命中率由上而下排序

```python
data6 = df[["period","shot_made_flag","shot_type","game_date","opponent","game_id"]]
gameid = list(set(data6["game_id"]))
first_half = {}
second_half ={}


for i,data in data6.iterrows():
    if(data["shot_made_flag"]==1):
        if(data["period"]==1 or data["period"]==2):
            if(data["game_id"] in first_half):
                if(data["shot_type"]=="2PT Field Goal"):
                    first_half[data["game_id"]]+=2
                elif(data["shot_type"]=="3PT Field Goal"):
                    first_half[data["game_id"]]+=3
            else:
                if(data["shot_type"]=="2PT Field Goal"):
                    first_half[data["game_id"]]=2
                elif(data["shot_type"]=="3PT Field Goal"):
                    first_half[data["game_id"]]=3
        elif(data["period"]==3 or data["period"]==4):
            if(data["game_id"] in second_half):
                if(data["shot_type"]=="2PT Field Goal"):
                    second_half[data["game_id"]]+=2
                elif(data["shot_type"]=="3PT Field Goal"):
                    second_half[data["game_id"]]+=3
            else:
                if(data["shot_type"]=="2PT Field Goal"):
                    second_half[data["game_id"]]=2
                elif(data["shot_type"]=="3PT Field Goal"):
                    second_half[data["game_id"]]=3
    
        
accurate = df.groupby(by = "game_id")["shot_made_flag"].mean()
```

```python
game2 = []
for i in gameid:
    if(i not in first_half):
        first_half[i]=0
    if(i not in second_half):
        second_half[i]=0
    oppo = list(set(data6[data6["game_id"]==i]['opponent']))
    if (first_half[i] > second_half[i]):
        game2.append([i,accurate[i],(first_half[i]-second_half[i]),(first_half[i]+second_half[i]),oppo[0]])
```

```python
game2.sort(key = lambda p: p[1],reverse = False)
game2[:3]
```

計算 Kobe「投籃連續失手最多球」之前 3 名場次，列出日期、對手、連續失手球數、
及該場得分，按失手次數由上而下排序

```python
data7 = df[["game_id","shot_made_flag","opponent","shot_type"]]
data7
result7=[]
for game_id in gameid:
    oppo2 = list(set(data6[data6["game_id"]==game_id]['opponent']))[0]
    flag=0
    maxflag =0
    point = 2*len(data7[data7["game_id"].isin([game_id]) & data7["shot_made_flag"].isin([1]) & data7["shot_type"].isin(["2PT Field Goal"])])+3*len(data7[data7["game_id"].isin([game_id]) & data7["shot_made_flag"].isin([1]) & data7["shot_type"].isin(["3PT Field Goal"])])
    
    for index, data in data7[data7["game_id"]==game_id].iterrows():
        if(data["shot_made_flag"]==0):
            flag+=1
        elif(data["shot_made_flag"]==1):
            if(flag>maxflag):
                maxflag=flag
            flag=0
    result7.append([game_id,maxflag,oppo2,point])
```

```python
result7.sort(key = lambda p: p[1],reverse = True)
result7[:3]
```
